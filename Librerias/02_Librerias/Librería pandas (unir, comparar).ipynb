```python
import pandas as pd
```

```python
#  Datos de prueba
files_1 = pd.DataFrame([["20/02/2019", "RutrumAc.gif", 219, "bb3f4d720a56c1fea06f8ed61d2d280d"],
                        ["20/02/2019", "DuiVel.avi", 233, "d0487dd6ac871f534f27b751ab1d19b0"],
                        ["06/01/2019", "DisParturient.tiff", 184, "d3cc5726c760c084e9ae5d96ab77abc8"],
                        ["22/02/2019", "SedAccumsanFelis.png", 128, "0035dc0b36335f72daf37e3ee249ae3a"],
                        ["11/02/2019", "Id.jpeg", 106, "b30d21f70a7a243a3d2d75d0d75096f2"],
                        ["29/01/2019", "RidiculusMus.mp3", 64, "4d14558d61f8bdb82a1bd88f3e452b94"]],
                       columns = ["date", "file", "size", "md5"])
files_2 = pd.DataFrame([["20/02/2019", "RutrumAc.gif", 219, "bb3f4d720a56c1fea06f8ed61d2d280d"],
                        ["20/02/2019", "DuiVel.avi", 89,"2f01fa4808a3fa0c5e71401d269f2aa8"],
                        ["29/01/2019", "RidiculusMus.mp3", 64, "4d14558d61f8bdb82a1bd88f3e452b94"],
                        ["11/01/2019", "Sapien.jpeg", 140, "22a2441ec32b97d86a7ef8f2e8d747d2"],
                        ["26/02/2019", "SemMauris.mp3", 169, "be4c1736150634f27e2ed956a9b1a1d0"],
                        ["07/02/2019", "EstPhasellusSit.tiff", 194, "8df4bd43d73df7afcb96d3d13a3fcb44"],
                        ["06/01/2019", "DisParturient.tiff", 184, "d3cc5726c760c084e9ae5d96ab77abc8"],
                        ["22/02/2019", "SedAccumsanFelis.png", 128, "0035dc0b36335f72daf37e3ee249ae3a"]],
                       columns = ["date", "file", "size", "md5"])
```

### Unión de dataframe **merge**

Método de unión **how**: **left**, **right**, **outer** o **inner** (comúnes en los dos)

```python
#  Obtener registros comunes en los dos
files_1.merge(files_2, how='inner')
```

### Unión de todos y sacar columna con los "cambios" (*indicator**)

```python
#  Unimos los dos dataframes, usamos outer para coger los comunes y los que no
#  y obtenemos una columna 'union' que indica si existe en los dos, solo en la derecha o sólo en la izquierda
files_1.merge(files_2, how='outer', indicator='union')
```

```python
# Y nos quedamos sólo con los de la izquierda
df = files_1.merge(files_2, how='outer', indicator='union')
df[df.union=='left_only']
```

### Comparar registros de dos Dataframes

```python
df1 = pd.DataFrame({'jugador': ['A', 'B', 'C', 'D'],
                    'puntos': [12, 15, 17, 24],
                    'asiste': [4, 6, 7, 8]})
df1
```

```python
df2 = pd.DataFrame({'jugador': ['A', 'B', 'C', 'D'],
                    'puntos': [12, 24, 26, 29],
                    'asiste': [7, 8, 10, 13]})
df2
```

### Comparar dos DataFrame, son iguales, equals

```python
#  ver si dos DataFrames son idénticos 
df1.equals(df2)
```

### Comparar DataFrame restando los valores, **subtract**

```python
# Restar df1 de df2 
df2.set_index('jugador').subtract(df1.set_index ('jugador'))
```

### Fusionar con merge outer e indicator

```python
# outer fusiona los dos DataFrames, agregando una columna indicadora llamada 'Exist' 
diff_df = pd.merge(df1, df2, how= 'outer', indicator = 'Existir')

# encontrar qué filas no existen en ambos DataFrames
diff_df = diff_df.loc[diff_df['Existir'] != 'both']
diff_df
```

### Uso **compare** para comparar dos DataFrame

```python
#  Alinear las diferencias en las columnas
df1.compare(df2)
```

```python
#  Alinear diferencias en filas
df1.compare(df2, align_axis=0)
```

```python
#  Mantener valores iguales
df1.compare(df2, keep_equal=True, align_axis=0)
```

```python
#  Mantener todas las filas y columnas originales y también todos los valores originales
df1.compare(df2, keep_shape=True, keep_equal=True)
```
