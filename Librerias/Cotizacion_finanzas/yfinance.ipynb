````python
#   Ejemplo 1 - Descargar historial de un activo

import yfinance as yf


#   Descargar el historial de Apple (AAPL) desde el 1 enero 2024 al 8 de marzo 2024
aapl = yf.download('AAPL', start='2024-01-01', end='2024-03-09')

#   Imprimimos los primeros registros del dataframe
print(aapl.head())
````

````python
#   Ejemplo 2 - Informaci칩n en tiempo real

import yfinance as yf

aapl_quote = yf.Ticker('AAPL').info

print(aapl_quote)
````

````python
#   Ejemplo 3 - Descarga Multiple de datos de activos

import yfinance as yf

activos = ['AAPL', 'MSFT']
data = yf.download(activos, start='2024-01-01', end='2024-03-09')

print(data.head())
````

````python
#   Ejemplo 4 - An치lisis de datos financieros

import yfinance as yf
import matplotlib.pyplot as plt

#   Descargamos el historial de datos de Microsot
msft = yf.download('MSFT', start='2024-01-01', end='2024-03-09')

#   Representamos los precios de cierre
plt.figure(figsize=(10, 6))
msft['Close'].plot(title='Microsoft precios de cierre')
plt.xlabel('Fecha')
plt.ylabel('Precios de cierre')
plt.show()
````

````python
#   Ejemplo 5 - Splits y Dividendos

import yfinance as yf

aapl_dividendos = yf.Ticker('AAPL').dividends
aapl_splits = yf.Ticker('AAPL').splits

print('Dividendos\n', aapl_dividendos)
print("\nSplits\n", aapl_splits)
````
