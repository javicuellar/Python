### Ejemplo 1.- recuperar todo el código fuente HTML de la página

1) Configurar webdriver, puede ser Chrome, Firefox, ...
    - driver = webdriver.Chrome()
2) Leer página web
    - driver.get('https://www.example.com')
3) Obtener código HTML de la página
    - html_source = driver.page_source
4) Cerrar driver
    - driver.quit()

```python
# Función para obtener el código fuente HTML de la página
def Obtener_HTML(url, opcion="pantalla"):
    # Configurar el WebDriver (usando Chrome en este ejemplo) 
    driver = webdriver.Chrome() 
    
    # Navegar a la página web deseada 
    driver.get(url) 

    # Obtener el código fuente HTML de la página 
    html_source = driver.page_source 

    if opcion == "pantalla":
        print(html_source)
    elif opcion == "fichero":
        f = open("web.html", "w")
        f.write(html_source)
        f.close()
    
    # Cerrar el WebDriver 
    driver.quit()

    return html_source
```

```python
url = 'https://www.example.com'

# Navegar a la página web deseada 
# Obtener_HTML(driver, url, "pantalla")
Obtener_HTML(url, "fichero")
```

### Buscar campo en web, informar con datos y pulsar botón

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
# Función para realizar búsqueda en el buscador Duckduckgo
def buscar_duckduckgo(termino, opcion="pantalla"):
    chrome_options = Options()
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                                "AppleWebkit/537.36 (KHTML, like Gecko) "
                                "Chrome/91.0.4472.77 Safari/537.36")

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)

    driver.get("https://duckduckgo.com")

    # Encontrar termino en caja de búsqueda
    caja_busqueda = driver.find_element(By.NAME, "q")
    caja_busqueda.send_keys(termino)
    caja_busqueda.send_keys(Keys.RETURN)

    # Esperar a que se carguen los resultados
    import time, random
    time.sleep(random.randint(2,5))
    
    # Obtener el código fuente HTML de la página 
    html_source = driver.page_source 

    import json
    json_data = json.dumps(html_source, indent=4)

    if opcion == "pantalla":
        print(json_data)
    elif opcion == "fichero":
        f = open("web.html", "w")
        f.write(html_source)
        f.close()
    
    # no Cerramos para devolver el WebDriver 
    # driver.quit()
    return driver
```

### Ejemplo 2.- Realizar búsqueda de terminos en el buscador DuckDuckgo

```python
# Función para realizar búsqueda en el buscador Duckduckgo
def buscar_duckduckgo(termino):
    chrome_options = Options()
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                                "AppleWebkit/537.36 (KHTML, like Gecko) "
                                "Chrome/91.0.4472.77 Safari/537.36")

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)

    import time,random
    time.sleep(random.randint(1,5))     # esperar entre 1 y 5 segundos, simular ser humano

    driver.get("https://duckduckgo.com")
    driver.maximize_window()

    # Encontrar termino en caja de búsqueda
    caja_busqueda = driver.find_element(By.NAME, "q")
    caja_busqueda.send_keys(termino)
    caja_busqueda.send_keys(Keys.RETURN)

    # Esperar a que se carguen los resultados
    time.sleep(random.randint(2,5))
        
    # Extraer títulos y enlaces de los resultados
    # resultados = driver.find_elements(By.CSS_SELECTOR, "h2 a")
    resultados = driver.find_elements(By.TAG_NAME, "a")

    salida = []
    for resultado in resultados:
        try:
            titulo = resultado.text                     # Texto del enlace (título del resultado)
            enlace = resultado.get_attribute("href")     # Enlace del resultado
            if titulo and enlace:
                    salida.append({
                        "Término": termino,
                        "Título": titulo,
                        "Enlace": enlace
                        })
        except Exception:
            continue
 
    driver.quit()

    return salida
```

```python
# Búsqueda de los términos indicados en duckduckgo
terminos_busqueda = "automatización con Python"
buscar_duckduckgo(terminos_busqueda)
```
